```python
import cv2
import mediapipe as mp
from function import getCenter, getCollidePoint, getValue, getPoint, whetherPassFromYaxis, filterPass, whetherPassFromXaxis
import matplotlib.pyplot as plt
#define
infinity = 100000
color_green = (0, 255, 0)
color_red = (0, 0, 255)
color_blue = (255, 0, 0)
const_rate = 4 #尋找手臂斜率及寬度的比例
K = 3
xaxis_grid = 10
yaxis_grid = K * xaxis_grid
###
mpHands = mp.solutions.hands
hands = mpHands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5)
mpDraw = mp.solutions.drawing_utils
handLmsStyle = mpDraw.DrawingSpec(color = color_blue, thickness=3)
handConStyle = mpDraw.DrawingSpec(color = color_green, thickness=5)

img = cv2.imread('label/3.PNG')
result = hands.process(img)

#二值化描邊
img = cv2.Canny(img, 60, 256)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#二值化
#ret, img = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
####

if result.multi_hand_landmarks:
    imgHeight = img.shape[0]
    imgWidth = img.shape[1]
    originx = 0 #腕關節
    originy = 0

    for handLms in result.multi_hand_landmarks: 
        for i, lm in enumerate(handLms.landmark):
            xPos = round(lm.x * imgWidth)
            yPos = round(lm.y * imgHeight)
            if i == 0:
                originx = xPos
                originy = yPos
        

m_x, m_y = getCenter(originx, round(originy + (imgHeight / const_rate)), img) #找斜率用的x和y

if m_x == originx:
    print("infinity")
    ydir_m = infinity
else:
    ydir_m = -1 * (originy - m_y) / (originx - m_x)

xdir_m = -1 / ydir_m

xdirValue = getValue(m_x, m_y, ydir_m, img)/2
ydirValue = round(K * xdirValue)
xdirValue = round(xdirValue)
xdirEnd_x, xdirEnd_y = getPoint(originx, originy, xdir_m,  xdirValue, result.multi_handedness[0].classification[0].label, False,img)
ydirEnd_x, ydirEnd_y = getPoint(originx, originy, ydir_m,  ydirValue, result.multi_handedness[0].classification[0].label, True,img)
xdirValue/10 #分為10部分, 每部份的中間點, 1/20, 3/20, 5/20....
xdir_point = [[0 for _ in range(2)] for _ in range(xaxis_grid)]#每個點有x和y
ydir_point = [[0 for _ in range(2)] for _ in range(yaxis_grid)]#每個點有x和y
for i in range(xaxis_grid):
    xdir_point[i][0], xdir_point[i][1] = getPoint(originx, originy, xdir_m, ((i/xaxis_grid + 1/(xaxis_grid * 2)) * xdirValue), result.multi_handedness[0].classification[0].label, False,img)
for i in range(yaxis_grid):
    ydir_point[i][0], ydir_point[i][1] = getPoint(originx, originy, ydir_m, ((i/yaxis_grid + 1/(yaxis_grid *2)) * ydirValue), result.multi_handedness[0].classification[0].label, True,img)
whethery = [0] * K * xaxis_grid #紀錄每個點所碰到的最近距離, 0表示沒有成功通過兩個點以上
for i in range(K * xaxis_grid):
    whethery[i] = whetherPassFromYaxis(ydir_point[i][0], ydir_point[i][1], xdir_m, result.multi_handedness[0].classification[0].label, xdirValue, img)[0]
whetherx = [0] * 
for i in range(xdirValue):
    
whethery = filterPass(whethery, xdirValue / 6)
###############################print################################################
cv2.circle(img, (originx, originy), 10, color_red, -1)   
cv2.circle(img, (xdirEnd_x, xdirEnd_y), 10, color_red, -1)
cv2.line(img, (originx, originy), (xdirEnd_x, xdirEnd_y), color_red, 5)
cv2.circle(img, (ydirEnd_x, ydirEnd_y), 10, color_red, -1)
cv2.line(img, (originx, originy), (ydirEnd_x, ydirEnd_y), color_red, 5)
cv2.circle(img, (m_x, m_y), 10, color_blue, -1)          #用這條來顯示當前選擇的手臂粗度位置
#cv2.line(img, (originx, originy), (m_x, m_y), color_red, 5)
if result.multi_handedness[0].classification[0].label == "Left":# hand type
    print("hand type: right")
else: 
    print("hand type: left")
print("ydir_m: ", ydir_m, ", xdir_m: ", xdir_m)#m for direction
print("xdir_value: ", xdirValue)#axis length
#print("xaxis_point:")
for i in range(xaxis_grid):
    #print("(", xdir_point[i][0], ", ", xdir_point[i][1], ")")
    cv2.circle(img, (xdir_point[i][0], xdir_point[i][1]), 4, color_blue, -1)
#print("yaxis_point:")
for i in range(yaxis_grid):
    #print("(", ydir_point[i][0], ", ", ydir_point[i][1], ")")
    cv2.circle(img, (ydir_point[i][0], ydir_point[i][1]), 4, color_blue, -1)
print("first_touch:")
for i in range(K * xaxis_grid):
    print(whethery[i])
#another side point
another_xdir_point = [[0 for _ in range(2)] for _ in range(xaxis_grid)]#每個點有x和y
another_ydir_point = [[0 for _ in range(2)] for _ in range(yaxis_grid)]#每個點有x和y
for i in range(xaxis_grid):
    another_xdir_point[i][0], another_xdir_point[i][1] = getPoint(xdir_point[i][0], xdir_point[i][1], ydir_m, ydirValue, result.multi_handedness[0].classification[0].label, True,img)
for i in range(yaxis_grid):
    another_ydir_point[i][0], another_ydir_point[i][1] = getPoint(ydir_point[i][0], ydir_point[i][1], xdir_m, xdirValue, result.multi_handedness[0].classification[0].label, False,img)
for i in range(xaxis_grid):
    cv2.line(img, (xdir_point[i][0], xdir_point[i][1]), (another_xdir_point[i][0], another_xdir_point[i][1]), color_red, 1)
    #cv2.circle(img, (another_xdir_point[i][0], another_xdir_point[i][1]), 4, color_blue, -1)
for i in range(yaxis_grid):
    cv2.line(img, (ydir_point[i][0], ydir_point[i][1]), (another_ydir_point[i][0], another_ydir_point[i][1]), color_red, 1)
    #cv2.circle(img, (another_ydir_point[i][0], another_ydir_point[i][1]), 4, color_blue, -1)

mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS, handLmsStyle, handConStyle)
img = cv2.resize(img, None, fx = 0.5, fy = 0.5, interpolation = cv2.INTER_CUBIC)
cv2.imshow('img', img)    
cv2.waitKey(0)
cv2.destroyAllWindows()
#plt.imshow(img)
```

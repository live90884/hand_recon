```python
import matplotlib.pyplot as plt
import cv2, os, re
from PIL import Image
from numpy import expand_dims
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array

path = 'data'
imgNum = 300 #total photo
datagen = ImageDataGenerator(rotation_range = 40, 
                            width_shift_range = 0.2, 
                            height_shift_range = 0.2, 
                            zoom_range = 0.1, 
                            horizontal_flip = True,
                            shear_range = 0.2, 
                            fill_mode = 'nearest',
                            brightness_range = [0.2, 1.0])
for i, folderName in enumerate(os.listdir(path)):
    print("folder:", folderName)
    for file_name in os.listdir(os.path.join(path, folderName)):
        if file_name == "train":
            count = 0
            for image in os.listdir(os.path.join(path, folderName, file_name)):
                count+=1
            if count >= imgNum:
                continue
            count = imgNum - count
            savepath = os.path.join(path, folderName, file_name)
            alter = 0
            for image in os.listdir(os.path.join(path, folderName, file_name)):
                print(count)
                img = cv2.imread(os.path.join(path, folderName, file_name, image))
                data = img_to_array(img)
                samples = expand_dims(data, 0)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = img.reshape((1,) + img.shape)
                i = 0
                
                for batch in datagen.flow(img, batch_size = 10, save_to_dir = savepath, save_prefix = 'hand',save_format = 'png'):
                    augImage = batch[0]
                    augImage = augImage.astype('float32')
                    augImage /= 255
                    i += 1
                    count -= 1
                    if count <= 0:
                        alter = 1
                        break
                    if i >= 19:
                        break
                
                if alter == 1:
                    break
```
